# **Outlier's Detection**

```python

numerical_features = ['temp', 'atemp', 'humidity', 'windspeed', 'count']

Q1 = bike_data[numerical_features].quantile(0.25)
Q3 = bike_data[numerical_features].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((bike_data[numerical_features] < lower_bound) | (bike_data[numerical_features] > upper_bound)).any(axis=1)

bike_data_cleaned = bike_data[~outliers]

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=bike_data[numerical_features])
plt.title('Boxplot of Numerical Features (Before)')
plt.subplot(1, 2, 2)
sns.boxplot(data=bike_data_cleaned[numerical_features])
plt.title('Boxplot of Numerical Features (After)')
plt.tight_layout()
plt.show()
```

# **Check if there any significant difference between the no. of bike rides on Weekdays and Weekends?**

```python
# H0 = Average bike rides on Weekdays and weekends are same.
# H1= Average bike rides on Weekdays and weekends are different.
# alpha= 0.05
weekends=df[(df['workingday']==0)]['count']
weekdays=df[df['workingday']==1]['count']
```

```python
from scipy.stats import ttest_ind
ttest_ind(weekends,weekdays)
```

```python
weekends.sum(),weekdays.sum()
```

```python
df['workingday'].value_counts()
```

P-value=0.226 which is greater than the significant value(0.05), so we accept the null hypothesis.
Hence there is no significant differnece between bike rides on weekdays and weekends.


#**Check if the demand of bicycles on rent is the same for different Weather conditions?**

**Histogram for different Weather**

```python
weather=['clear','cloudy','lightrain','heavyrain']
for i,features in enumerate(weather,start=1):
  plt.subplot(2,2,i)
  sns.histplot(bike_data[bike_data['weather']==i]['count'],kde=True)
  plt.title(f"Histplot of {features}")
  plt.xlabel(features)
plt.tight_layout()
plt.show()
```

```python
df['weather'].value_counts()
```

```python
clear=df[df['weather']==1]['count']
cloudy=df[df['weather']==2]['count']
lightrain=df[df['weather']==3]['count']
heavyrain=df[df['weather']==4]['count']
```

**QQ Plot for Different Weather**

```python
import statsmodels.api as sm
for i,features in enumerate(weather,start=1):
  plt.subplot(2,2,i)
  sm.qqplot(bike_data[bike_data['season']==i]['count'],line='s',ax=plt.gca())
  plt.title(f"{features}")
plt.tight_layout()
plt.show()
```

Hence we can see that the data is not gaussian

**Skewness and Kurtosis for Different Weather Conditions**

```python
from scipy.stats import skew,kurtosis
weather_list=[clear,cloudy,lightrain,heavyrain]
print(f"Skewness and Kurtosis for Clear are {skew(weather_list[0])},{kurtosis(weather_list[0])}")
print(f"Skewness and Kurtosis for Cloudy are {skew(weather_list[1])},{kurtosis(weather_list[1])}")
print(f"Skewness and Kurtosis for Lightrain are {skew(weather_list[2])},{kurtosis(weather_list[2])}")
print(f"Skewness and Kurtosis for Heavyrain are {skew(weather_list[3])},{kurtosis(weather_list[3])}")
```

**Shapiro-Wilk Test**

```python
# H0= Data is Gaussian
# H1= Data is not Gaussian
from scipy.stats import shapiro
print(f"for Clear  {shapiro(weather_list[0])}")
print(f"for Cloudy   {shapiro(weather_list[1])}")
print(f"for Lightrain   {shapiro(weather_list[2])}")
print(f"for Heavyrain    Not Enough Data")
```

**Levene Test**

```python
# used to check if there is equal variance in different groups
# H0= All the groups have equal variance
# H1= Have different variance
from scipy.stats import levene
levene(clear,cloudy,lightrain,heavyrain)
```

Hence we reject null hypothesis, means that there is different variance among different groups

**All the Assumptions of One-Way Anova test failed. therefore we will do KRUSKAL TEST**

**KRUSKAL TEST**

```python
from scipy.stats import kruskal
kruskal(clear,cloudy,lightrain,heavyrain)
```

p-value is very less than significant value i.e 0.05. Hence we reject null Hypothesis. Therefor we can conclude that the average ride on different weather is different

# **Check if the demand of bicycles on rent is the same for different Seasons?**

**HISTOGRAM OF DIFFERENT SEASONS**

```python
season=['spring','summer','fall','winter']
for i,features in enumerate(season,start=1):
  plt.subplot(2,2,i)
  sns.histplot(bike_data[bike_data['season']==i]['count'],kde=True)
  plt.title(f"Histplot of {features}")
plt.tightlayout()
plt.show()
```

**QQ PLOT OF DIFFERENT SEASONS**

```python
import statsmodels.api as sm
season=['spring','summer','fall','winter']
for i,features in enumerate(season,start=1):
  plt.subplot(2,2,i)
  sm.qqplot(bike_data[bike_data['season']==i]['count'],line='s',ax=plt.gca())
  plt.title(f"{features}")
plt.tight_layout()
plt.show()
```

from the figures we found out that the given data is not normally distributed

**Skewness and Kurtosis Test**

```python
spring=df[df['season']==1]['count']
summer=df[df['season']==2]['count']
fall=df[df['season']==3]['count']
winter=df[df['season']==4]['count']
```

```python
from scipy.stats import skew,kurtosis
season_list=[spring,summer,fall,winter]
skew(season_list[0]),kurtosis(season_list[0])
```

```python
print(f"Skewness and Kurtosis for Spring are {skew(season_list[0])},{kurtosis(season_list[0])}")
print(f"Skewness and Kurtosis for Summer are {skew(season_list[1])},{kurtosis(season_list[1])}")
print(f"Skewness and Kurtosis for Fall are {skew(season_list[2])},{kurtosis(season_list[2])}")
print(f"Skewness and Kurtosis for Winter are {skew(season_list[3])},{kurtosis(season_list[3])}")
```

**Shapiro-Wilk Test**

```python
from scipy.stats import shapiro
# H0= Data is Gaussian
# H1= Data is not Gaussian
```

```python
print(f"for SPRING   {shapiro(season_list[0])}")
print(f"for SUMMER   {shapiro(season_list[1])}")
print(f"for FALL   {shapiro(season_list[2])}")
print(f"for WINTER   {shapiro(season_list[3])}")
```

Hence we reject the null hypothesis for all the Seasons and can say that data is not normal

**LEVENE'S TEST**

```python
# used to check if there is equal variance in different groups
# H0= All the groups have equal variance
# H1= Have different variance
```

```python
from scipy.stats import levene
levene(season_list[0],season_list[1],season_list[2],season_list[3])
```

Hence we reject null hypothesis, means that there is different  variance among different groups


**All the Assumptions of One-Way Anova test failed. therefore we will do KRUSKAL TEST**

**KRUSKAL-WALLI'S TEST**

```python
# H0= Population median of all group is same
# H1= At least one median is different
from scipy.stats import kruskal
kruskal(season_list[0],season_list[1],season_list[2],season_list[3])
```

p-value is very less than significant value i.e 0.05. Hence we reject null Hypothesis

# Check if the Weather conditions are significantly different during different Seasons?

```python
# H0= There is no significant association between weather conditions and seasons.
# H1= There is a significant association between weather conditions and seasons.
weather_season=pd.crosstab(df['weather'],df['season'])
weather_season
```

```python
from scipy.stats import chi2_contingency
chi2_contingency(weather_season)
```

P-value is very low, so we can reject null hypothesis. hence we can say that there is a significant assosiation between weather and season
